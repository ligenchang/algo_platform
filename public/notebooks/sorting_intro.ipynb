## Bubble Sort Explained
Bubble Sort is a simple comparison-based algorithm.

```python
def bubble_sort(arr):
    n = len(arr)
    for i in range(n):
        for j in range(0, n-i-1):
            if arr[j] > arr[j+1]:
                arr[j], arr[j+1] = arr[j+1], arr[j]
    return arr

bubble_sort([5, 2, 9, 1, 5, 6])
```
